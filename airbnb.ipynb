```python
x= df['numberOfGuests']
y= df['pricing']
x_mean = x.mean()
y_mean = y.mean()
x_mean,y_mean
Sxy = np.sum((x - x_mean) * (y - y_mean))
Sxx = np.sum(np.square(x - x_mean))
Sxy,Sxx
b1 = Sxy/Sxx
b0 = y_mean - (b1 * x_mean)
print('Slope b1 is ',b1)
print('Intercept b0 is',b0)
def line(x):
    return b1*x + b0
plt.scatter(x,y)
plt.plot(x,line(x), color="red")
plt.xlabel('numberOfGuests')
plt.ylabel('pricing')
```
